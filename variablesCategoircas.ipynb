```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.tail(2)
```

```python
sns.countplot(data=tips, x = 'day', hue='sex')
```

### diagramas categóricos

```python
plt.figure(figsize= (6,6))
sns.stripplot(data = tips,x ='day',y ='total_bill', hue='sex', dodge= True)
```

```python
plt.figure(figsize= (6,6))
sns.swarmplot(data = tips, dodge=True,x ='day',y ='total_bill' )
plt.show()
```

```python
sns.boxplot(data= tips, x ='day',y ='total_bill', hue='sex')
```

```python
plt.figure(figsize= (6,6))
sns.boxplot(data= tips, x ='day',y ='total_bill', hue='sex')

sns.swarmplot(data = tips, dodge=True,x ='day',y ='total_bill' )
plt.show()
```

```python
sns.violinplot(data= tips,x ='day',y ='total_bill')
```

```python
sns.catplot(data = tips,x ='day',y ='total_bill', col= 'time')
```

### Scaterplot

```python
markers = { "Dinner":"s" }
sns.scatterplot(data = tips,x='total_bill', y='tip', style='day', hue='day',size='size', markers=markers)
plt.legend(loc = 'center', bbox_to_anchor =(1.12,0.5))
plt.show()
```

```python
sns.lineplot( data = tips, x='total_bill',y='tip', hue='day',style='time',size='time')
```
