### parámeotrs usados con seabornn


```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data = tips, x= 'total_bill', y = 'tip', legend=True, hue= 'sex')
```

```python
sns.displot(data = tips, x= 'total_bill',  legend=True, hue= 'sex', kind = 'kde', palette= 'dark', alpha = 0.75)
```

### distribuciones

```python
tips.head()
```

```python
sns.histplot(data = tips, x = 'tip', bins= 15, stat= 'percent')
```

```python
sns.histplot(data = tips, x = 'tip', bins= 15, cumulative=False,hue='sex', stat= 'probability', multiple='dodge')
```

```python
sns.kdeplot(data = tips,x='total_bill', hue= 'sex')
```

```python
sns.kdeplot(data = tips,x='total_bill', hue= 'sex', cumulative=True, shade=True)
```

```python
sns.kdeplot(data = tips,x='total_bill', hue= 'sex',bw_adjust=1)
```

```python
sns.displot(data = tips,y ='smoker', x = 'total_bill', kind='hist')
```
