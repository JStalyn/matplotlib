```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python


fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
# posición tamaño  
axes.plot(x,y, 'b')
plt.show()
```

```python
figure = plt.figure()
axes = figure.add_axes([0.5,0.4,0.8,0.9])
axes02 = figure.add_axes([0.7,0.8,0.2,0.2])
axes.plot(x,y,'b')

axes02.plot(y,x,'r')
axes02.set_facecolor('blue')
plt.show() 
```

SUBPLOTs

```python
import matplotlib.pyplot as plt
import numpy as np  
x = np.linspace(0,10,20)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols =2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, ((axes1, axes2),(axes3, axes4)) = plt.subplots(nrows = 2, ncols =2)
axes1.plot(x,y)
axes2.plot(np.cos(y),x)
axes3.plot(np.sin(x),y)
axes4.plot(np.tan(y),x)
```
