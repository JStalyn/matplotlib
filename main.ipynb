```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.plot(y,x)
plt.subplot(1,2,2)
plt.hist(y)
plt.hist(x)
plt.show
```
